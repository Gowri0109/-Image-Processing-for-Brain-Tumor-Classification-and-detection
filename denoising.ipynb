```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd 
```

```python
#To check the type of noise present
```

```python
# Set the directory containing the images
img_dir = "D:\master_thesis\datasets\grayscale_glioma"

# Loop through all the images in the directory
for filename in os.listdir(img_dir):
    # Read the image
    img_path = os.path.join(img_dir, filename)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

    # Calculate the histogram of the image
    hist = cv2.calcHist([img], [0], None, [256], [0, 256])

    # Determine the type of noise based on the shape of the histogram
    std_dev = np.std(hist)
    if std_dev < 1000:
        print(filename + ': Salt-and-Pepper Noise')
    elif std_dev > 3000:
        print(filename + ': Gaussian Noise')
    else:
        print(filename + ': Unknown Noise')
```

```python
# To choose the kernel size
```

```python

# Define folder path
folder_path = 'D:\master_thesis\datasets\grayscale_glioma'

# Loop over all images in folder
for file_name in os.listdir(folder_path):
    # Read image
    image = cv2.imread(os.path.join(folder_path, file_name))
    
    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Initialize variables for best PSNR and corresponding kernel size
    best_psnr = 0
    best_kernel_size = None
    
    # Loop over different kernel sizes
    for k in range(3, 22, 2):
        # Apply bilateral filter
        filtered = cv2.bilateralFilter(gray, k, 50, 50)
        
        # Calculate PSNR between original and filtered images
        psnr = cv2.PSNR(gray, filtered)
        
        # If current PSNR is better than the best so far, update variables
        if psnr > best_psnr:
            best_psnr = psnr
            best_kernel_size = k
    
    # Print optimal kernel size for current image
    print(f"Optimal kernel size for {file_name}: {best_kernel_size}")
```

```python
#denoising using multiplr filters
```

```python
# Set input and output folders
input_folder = "D:\master_thesis\datasets\grayscale_glioma"
output_folder = "D:\master_thesis\datasets\denoised\glioma"

# Set filter parameters
gaussian_kernel_size = 3
median_kernel_size = 3
bilateral_diameter = 9
bilateral_sigma_color = 75
bilateral_sigma_space = 75

# Loop through each image in the input folder
for file_name in os.listdir(input_folder):
    # Read the image
    image_path = os.path.join(input_folder, file_name)
    image = cv2.imread(image_path)

    # Apply filters to remove noise
    image = cv2.GaussianBlur(image, (gaussian_kernel_size, gaussian_kernel_size), 0)
    image = cv2.medianBlur(image, median_kernel_size)
    image = cv2.bilateralFilter(image, bilateral_diameter, bilateral_sigma_color, bilateral_sigma_space)

    # Save the output image to the output folder
    output_file_name = os.path.splitext(file_name)[0] + '_denoised.jpg'
    output_path = os.path.join(output_folder, output_file_name)
    cv2.imwrite(output_path, image)
```
